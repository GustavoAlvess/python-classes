```python
nacionalidade =  input("Digite sua nacionalidade (brasileiro, americano, canadense):".lower()) # solicitando nacionalidade e transformando todas as letras em minusculo
match nacionalidade :
   case "brasileiro" :
    print("Você é do Brasil!")
   case "americano" :
    print("Você é do Estados unidos!")
   case "canadense" :
    print("Você é do Canada!")
   case _:
    print("Nacionalidade não encontrada")
```
