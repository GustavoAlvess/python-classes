# *While*


```python
# %%time para ver o twmpo q o código demora pra rodar
n = int(input("Digite um número para calcular seu fatorial:"))

resultado = 1
i = 1

while i <= n:
  resultado = resultado * i
  i = i+1
print('O fatorial fatorial de {} é: {}'.format(n,resultado)) # SEM ESPAÇO para não dar varias rresposta
```

```python
''' ordem crescente '''
number = 1 # valor da variavel
while number <= 5 : # condição
  print(number) # ação
  number+= 1 # incrementar mais 1 e fazer o processo de novo
```

```python
''' ordem decrescente '''
number = 5 # valor da variavel
while number >= 1 : # condição
  print(number) # ação
  number-= 1 # incrementar mais 1 e fazer o processo de novo
```

```python
numero = int(input("Digite um número: "))

soma_pares = 0
soma_impares = 0
i = 1

while i <= numero:
  if i % 2 != 0:
    soma_impares = soma_impares + i
  i += 1
print("Somando os números ímpares de 1 até {} resulta em: {}".format(numero, soma_impares))
```
