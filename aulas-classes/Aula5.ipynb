```python
#declarando dicionário

person1 = {
    'nome': 'Gustavo' ,
    'sexo': ' Masculino' ,
    'idade': 16 ,
    'casado' : True
}

print(person1)
print(type(person1))

# usando função dict
person2 = dict(nome='Maria', sexo = 'Feminino' , idade = 28, casado=True)
print(person2)
print(type(person2))

print('-----------------------------------------')

#buscando pela chave do dicionário
print(person2['sexo'])
print(person2['idade'])

print('-----------------------------------------')
#criando um dicionario vazio
dict1 = {}
print(dict1)
dict1['chave 1'] = ' não está mais vazio!'
print(dict1)

print('-----------------------------------------')

dict2 = {'Pessoa 1': ['Gustavo', 16 , 'Professor'], 'Pessoa 2': ['Leitão' , 16 , 'Médica']}
print(dict2)
print(dict2['Pessoa 1'] [0])
```

```python
#dicionario aninhado

pessoa = {
    'nome': 'João',
    'idade': 16,
    'endereço': {
        'rua': 'Rua principal',
        'cidade': 'Cidade A',
        'estado': 'Estado x'
    }

}
print("Nome: ", pessoa['nome'])
print("Idade: ", pessoa['idade'])
print("Endereço: ", pessoa['endereço']['rua'],pessoa['endereço']['cidade'], pessoa['endereço']['estado'])
```

```python
pessoa.get('nome') # .get puxa o valor do campo pela chave
```

```python
#alterar a informação já criada
person1['casado'] = True
#adicionar um novo campo e chave
person1['Nacionalidade'] = 'Brasileiro'

print(person1)
print(type(person1))
```

```python
#para remover uma chave e  o valor associado de um ducionário, use o método pop
print(person1)
person1.pop('Nacionalidade')
print(person1)
```

```python
print(person1.values())# printa os valores dos campos
```

```python
#criando um dicionário
meu_dicionario = {'a': 1 ,'b': 2 ,'c': 3 ,}

#exibindo o dicionário original
print("Dicionário original:" , meu_dicionario)

#usando o método clear para remover os itens
meu_dicionario.clear

#exibindo dicionario após o clear
print("Dicionário após clear:", meu_dicionario)
```

```python
#deletando elemento do dicionario
del meu_dicionario['c']

print("Dicionário após deletar 'b' :" , meu_dicionario)
```

```python
#dicionário original

dicionario1 = { 'a': 1, 'b': 2}

#adicionando elementos de outro dicionario
dicionario2 = {'b': 3, 'c' : 4}
dicionario1.update(dicionario2)

#exibindo dicionario após atualização
print("Dicionário após update:", dicionario1)
```

```python
#estoque atual
estoque = {'maçãs': 10, 'bananas': 15, 'laranjas': 8}
print(f"Estoque atual: {estoque}")
#estoque modificado
estoque2 = {'maçãs': 8, 'bananas': 10 , 'laranjas': 6}
estoque.update(estoque2)
print("Estoque com preços alterados: " ,estoque)
```
