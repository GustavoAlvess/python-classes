# Aula 6 - Tupla

## tupla - (  )
### imutável

```python
#criação
frutas = ('banana' , 'uva', 'laranja')
print(frutas)
print(type(frutas))
```

```python
#verificar tamanho da tupla
len(frutas)
```

```python
# pegar primeiro e últio item da tupla
print(frutas[0])
print(frutas[-1])
```

```python
# provando que a tupla é imutável
'''
frutas.append('tomate') vai dar erro pq a tupla é imutável
'''
print(frutas)
```

```python
# você pode deletar a tupla, mas não o item dela
tupla_del = ('a':1 ,'d': 4)
print(tupla_del)

del tupla_del
print(tupla_del)
```

```python
#  lista tupla - list(tupla)
minha_tupla = ( 1, 2, 3)
lista_resultante = list(minha_tupla)
print(lista_resultante)
print(type(lista_resultante))
```

```python
# tupla lista - tuple(lista)
minha_lista= [4 ,5,6]
tupla_restaurante = tuple(minha_lista)
print(tupla_restaurante)
print(type(tupla_restaurante))
```

```python
# tupla com varios tipos de dados
tupla1 = (1,"Hello world!", True, 3.2, [2,4,5], (3,4,5))
print(tupla1)
```

# Atividade

```python
#criando a tupla
frutas = ('maçã' , 'abacaxi', 'caju')
fruta_adicional = list(frutas) # convertendo em lista
fruta_adicional.append('tomate') # adicionando novo item
fruta_adicional = tuple(fruta_adicional) #convertendo em tupla
print(fruta_adicional) # exibindo nova tupla
print(type(fruta_adicional)) # exibindo tipo da tupla
```
