# **Aula 4**

```python
#declarar uma lista
frutas = ['banana' , 'uva' , 'manga']

print(frutas)
print(type(frutas))

#posição na lista

print(frutas[0])
print('-------')
print(frutas[1])
```

```python
a_lista = ['banana' ,None,'uva' , 'manga', 4==4]
print(a_lista)


a_list = [23,None,3.14 , frutas, 3 <= 5]
print(a_list[3])
```

```python
# criando uma matriz 3x3

matriz = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matriz)
print(type(matriz))
#acessando elementos da lista
print(matriz[1][2])

#provando que lista é mutável
matriz_ = [
    [1,2,3],
    [4,5,'a'],
    [7,8,9]
]

print(matriz_)
print(type(matriz_))
#acessando elementos da lista
print(matriz_[1][2])

print('_________________')
print('')
#declrar lista vazia ( útil para adicionar os valores durante o código

lista_vazia = []
print(lista_vazia)
print('------------------')
print(len(lista_vazia))
print(len(frutas))

# pegar parte da lista
print('------------------')

print(a_lista)
print('\n')
print(a_lista[:2])
print('\n')
print(a_lista[2:4])
```

```python
#inserir elementos na lista

print(frutas) #lista atual
frutas.insert(0,(input('Digite uma fruta para adicionar a lista:')))#inserindo iten
print(frutas)# lista atualizada

#inserir no final
frutas.append(input('Digite uma fruta para adicionar a lista:'))#inserindo iten
print(frutas)

```

```python
#extendendo a lista
frutas_adicionais = ['caju', 'morango' , 'abacaxi']

frutas.extend(frutas_adicionais)
print(frutas)

'''
o extend acrescenta itens a lista, o append adiciona uma lista ao final da lista já criada
'''
```

```python
#criando lista
numeros = [1,2,3,4,5,6,3]

# removendo o número 3 da lista
numeros.remove(3) # remove o primeiro 3 que ele ver na lista
print(numeros)
```

```python
print(frutas)
alimento_removido = frutas.pop(5) #remove o iten e colca ela em uma variável
'''
se não tiver nenhum indice = pop() ele remove o último item da lista
'''
#remove o iten e colca ela em uma variável
print(frutas)
print(alimento_removido)
```

```python
print(frutas)
alimento_removido = frutas.pop()
print(frutas)
print(alimento_removido)
```

```python
#juntar listas com sinal de +
mais_frutas = frutas + ['morango'] + ['goiaba' , ' tangerina']
print(mais_frutas)
```

```python
#fazer uma copia da lista

mais_frutas_copia = mais_frutas.copy()
print(mais_frutas_copia)

#modificando lista
mais_frutas_copia.remove('goiaba')
mais_frutas_copia.pop()
print(mais_frutas_copia) # lista modificada
print(mais_frutas) # lista original não modificada
```
