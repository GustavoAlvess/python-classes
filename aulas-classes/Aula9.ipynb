# Lambda
- forma diferente de executar função

```python
# com função normal
def quadrado(x):
  return x**2

print(quadrado(5))
```

```python
# com lambda

quadrado = lambda x: x**2
print(quadrado(5)) # saída 25
```

```python
# com função

def soma(x,y):
  return x+y

print(soma(3,4))
```

```python
# lambda
soma = lambda x,y: x+y
print(soma(3,4))
```

```python
numeros = [1,2,3,4,5]
ao_quadrado = list(map(lambda x: x**2, numeros))
print(ao_quadrado)
```

```python
numeros = [1,2,3,4,5]
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
```
